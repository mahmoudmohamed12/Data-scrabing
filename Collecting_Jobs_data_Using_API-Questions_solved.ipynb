You can also view  the json file contents  from the following <a href = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json">json</a> URL.


```python
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json'
def get_number_of_jobs_T(technology):
    no_jobs = []
    r= requests.get(url)
    if r.ok:
        data = r.json()
        for i in range(len(data)):
            n = data[i]['Key Skills']
            if technology in n or technology.capitalize() in n or technology.lower() in n:
                no_jobs.append(1)
                
        number_of_jobs = len(no_jobs) 
    return technology,number_of_jobs
```

Create a python list of all locations for which you need to find the number of jobs postings.


```python
location=["Los Angeles", "New York", "San Francisco", "Washington DC", "Seattle"]
def get_number_of_jobs_Loc_list(locations):
    r=requests.get(url) 
    number_of_jobs_list=[]
    for loca in locations:
        
        number_jobs = []
        if r.ok:
            data=r.json()
            for i in range(len(data)):
                n = data[i]['Location']
                if n == loca:
                    
                    number_jobs.append(1)
        number_of_jobs = len(number_jobs)
        number_of_jobs_list.append({loca: number_of_jobs})
                    
    return number_of_jobs_list
```

Collect the number of job postings for the following languages using the API:

*   C
*   C#
*   C++
*   Java
*   JavaScript
*   Python
*   Scala
*   Oracle
*   SQL Server
*   MySQL Server
*   PostgreSQL
*   MongoDB


```python
import numpy as np
technologies = [' C ', ' C|', 'C#', 'c#', 'C++', 'c++', 'Java ', 'Java|', 'java|','java ', 'JAVA ', 'JAVA|', 'JavaScript', 'Javascript','javascript', 'Python', 'Scala', 'SCALA', 'scala', 'Oracle', 'oracle', \
                ' SQL Server', ' sql', ' SQL', '|sql', '|SQL', 'MySQL', 'mysql', 'Mysql','PostgreSQL', 'PostGreSQL','postgresql','MongoDB', 'mongodb', 'MongoDb']
country = ['Los Angeles', 'New York', 'San Francisco', 'Washington DC', 'Seattle', 'Austin', 'Detroit']
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json' 
r=requests.get(url)

def get_number_of_jobs_TL(technology, countries):
    number_jobs = np.zeros((len(technology),len(countries)))
    for tech in technology:
        for loca in countries: 
            if r.ok:
                data=r.json()
                for i in range(len(data)):
                    n = data[i]['Location']
                    t= data[i]['Key Skills']
                    if n == loca and tech in t :
                        q=tech
                        u=n
                        p = countries.index(u)
                        m = technology.index(q)
                        number_jobs[m][p]+=1
    return number_jobs
```

```python
data = get_number_of_jobs_TL(technologies, country)
df = pd.DataFrame(data=data ,index=technologies,columns=[ 'Los Angeles', 'New York', 'San Francisco', 'Washington DC', 'Seattle', 'Austin', 'Detroit'])
df
```
